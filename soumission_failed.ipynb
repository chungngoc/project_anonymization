```python
import numpy as np
import pandas as pd
import datetime as dt
import random
import string
import hashlib
import json
#Loading data
columns_name = ['id', 'date', 'latitude', 'longitude' ]
anon = pd.read_csv('test', names = columns_name, sep = '\t')

anon.head(50)
```

```python
df['date'] = anon['date']

df.head(50)
```

```python
df['date'] = df['date'].str.slice(stop = 16)
df['date'] = pd.to_datetime(df['date'], format = "%Y-%m-%d %H:%M")

#df['hour'] = df['date'].dt.hour
#df['day'] = df['date'].dt.day
df['week'] = df['date'].dt.isocalendar().week
```

```python
#Change date to random (be the same week)
def change_date(date):
    rand_day = int(np.random.choice([0,1], p=[0.999, 0.001]))
    return date - dt.timedelta(days=rand_day) if date.isocalendar()[1] == (date - dt.timedelta(days=rand_day)).isocalendar()[1] else date + dt.timedelta(days=rand_day)

```

```python
df.loc[df['id'] != 'DEL', 'date'] = df.loc[df['id'] != 'DEL', 'date'].apply(change_date)
```

```python
df['week'] = df['date'].dt.isocalendar().week
id_time = df.loc[df['id']!= 'DEL',['id', 'date', 'latitude', 'longitude','week']]
id_time['new_id'] =  str(random.randrange(10000,19999)) + id_time['id'].astype(str) + id_time['week'].astype(str) +str(random.randrange(10000,19999))

id_time['new_id'] = id_time['new_id'].apply(lambda x: hashlib.sha256(x.encode()).hexdigest())
df.loc[df['id'] != 'DEL', 'id'] = id_time['new_id']
```
