```python
#Methods

#Random minute
def random_minute(date):
    rand_min = random.randint(1,30)
    return date - dt.timedelta(minutes=rand_min) if date.hour == (date - dt.timedelta(minutes=rand_min)).hour else date + dt.timedelta(days=rand_min)

#Change date to random (be the same week)
def change_date(date):
    rand_day = random.choice([1,-1])
    rand_min = random.randint(1,59)
    return date - dt.timedelta(days=rand_day)+dt.timedelta(minutes=rand_min) if date.isocalendar()[1] == (date - dt.timedelta(days=rand_day) +dt.timedelta(minutes=rand_min)).isocalendar()[1] else date + dt.timedelta(days=rand_day) +dt.timedelta(minutes=rand_min)

#Change latitude and longitude to mean
def change_lat_long():       
    df['latitude'] =  df['latitude'].mean()
    df['latitude'] = df['latitude'].round(decimals=2)
    df['longitude'] =  df['longitude'].mean()
    df['longitude'] = df['longitude'].round(decimals=2)
```
